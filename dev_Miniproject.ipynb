```python
#correlation coffients
#correlation - It is a relationship between the two variables
#correlation coefficient - It shows the value of correlation coefficient
#There are three types of correlation coefficient 1. Pearson's r 2.Spearmann's rho 3.Kendall's tau

print(df[['Open','Close']].corr( method='pearson'))
#gross income and total are highly positively correlated
#Let us visualise graphically

plt.scatter(df['Open'],df['Close'])
plt.xlabel('Open')
plt.ylabel('Close')
plt.show()
```

```python
print(df[['Open']].corr( method='spearman'))
print(df[['Open','Low']].corr( method='kendall'))

#All the three types of correlation coefficients will have same value
```

```python
#Contigency tables - It is used to summarize the object datatypes
#It gives the numeric count of qualitative data in the data set
#The contigency table is acheived by giving the command pd.crosstab(), here the columns are represented as rows and rows are represented as index

contigency_table = pd.crosstab(index = df['Open'], columns = df['Close'])
contigency_table

#It is observed that most of Female customers prefer to pay using Cash
#And also it is observed that most of men prefer Ewallet as mode of payment
```

```python
#Margins can also be added to the contigency table which gives the total
contigency_table = pd.crosstab(index = df['Open'], columns = df['Close'], margins = True , margins_name = 'Total')
contigency_table
```

```python
#Calculating percentages
#Percentages can be computed using four different ways
# 1. Take the sum and multiply with 100
# 2. Using agg() function
# 3. Using apply() method
# 4. Using .div() method

# 1. Take the sum and multiply with 100
df['Total_percent'] = df['Total']/df['Total'].sum() * 100
df[['Total','Total_percent']]
```

```python
# 2. Using agg() function
data1 = df.groupby(['Open', 'Low']).agg({'Low': 'sum'})
data1
```

```python
# 3. Using apply() method
df3 = data1.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
df3
```

```python
# 4. Using .div() method
data2 = df.groupby(['Quantity', 'Total']).agg({'Quantity': 'sum'})
make = df.groupby(['Total']).agg({'Total': 'sum'})
data3=data2.div(make, level='Quantity') * 100
print(data3)
```
