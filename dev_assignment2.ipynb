```python
sns.lineplot(data=df,
             x='open',
             y='adjclose',
             legend='full')
#add title
plt.title('lineplot')
```

```python
sns.lineplot(data=df, 
             x='open', 
             y='volume', 
             legend='full')

# add title
plt.title('Seasonal plot')

# move the legend outside of the main figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2);
```

```python
sns.lineplot(data=df, 
             x='open', 
             y='high', 
             legend='full')

# add title
plt.title('Seasonal plot')

# move the legend outside of the main figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2);

# A seasonal trend is observed between Open stocks and High stocks
# The graph is noisy near the average of Open stocks
```

```python
#Lets see the correlation between these two

print(df['high'].corr(df['volume']))
```

```python
cols=3
rows=4
num_cols=df.select_dtypes(exclude='object').columns
fig=plt.figure(figsize=(cols*5,rows*5))
for i,col in enumerate(num_cols):
  ax=fig.add_subplot(rows,cols,i+1)
  sns.histplot(x=df[col],ax=ax)

fig.tight_layout()
plt.show()
```

```python
sns.lineplot(data=df, 
             x='high', 
             y='low', 
             legend='full')

# add title
plt.title('Seasonal plot')

# move the legend outside of the main figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2);
```
